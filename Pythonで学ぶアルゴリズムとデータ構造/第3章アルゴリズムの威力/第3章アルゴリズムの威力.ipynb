```
def simple_gcd(a, b):
  if a < b:
    a, b = b, a
  x = b
  while True:
    if a % b == 0 and b % x == 0:
      return x
    x -= 1
```

```
def euclidean_algorithm(a, b):
  while True:
    r = a % b
    if r == 0:
      return b
    a, b = b, r
```

```
def euclidian_recursion(a, b):
  r = a % b
  if r == 0:
    return b
  return euclidean_algorithm(b, r)
```

```
def selection_sort(array):
  x = array.copy()
  n = len(x)
  for i in range(n):
    min_idx = i
    for j in range(i, n):
      if x[j] < x[min_idx]:
        min_idx = j
    x[i], x[min_idx] = x[min_idx], x[i]
  return x
```

```
random.shuffle(my_array)
my_array
```

```
selection_sort(my_array)
```

```
import random

def rand_n_digit_int(n):
    return random.randint(10**(n-1), 10**n - 1)

rand_n_digit_int(100)


def euclidean_algorithm(a, b):
    while True:
        r = a % b
        if r == 0:
            return b
        a, b = b, r


a = rand_n_digit_int(100)
b = rand_n_digit_int(100)
euclidean_algorithm(a, b)


def simple_gcd(a, b):
    if a < b:
        a, b = b, a
    x = b
    while True:
        if a % x == 0 and b % x == 0:
            return x
        x -= 1

a = rand_n_digit_int(9)
b = rand_n_digit_int(9)
simple_gcd(a, b)
```

```
math.factorial(5)

def factorial(n):
  res = 1
  for i in range(n):
    res *= (i + 1)
  return res

math.factorial(5) == factorial(5)
```

```
def factorial_rec(n):
    if n == 1:
        return 1
    return n * factorial_rec(n-1)
factorial_rec(5)
```
