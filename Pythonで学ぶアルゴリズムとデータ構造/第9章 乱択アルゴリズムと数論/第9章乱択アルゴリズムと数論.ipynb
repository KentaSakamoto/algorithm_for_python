```
m = 3
# 3文字分のスペースを用意して右に詰めるテンプレートを用意
tmpl = '{:>3}' * 5
print(tmpl.format('a', 1, 2, 3, 4))
for a in range(m):
  print(tmpl.format(a, a % m, a**2 % m, a**3 % m, a**4 % m))
```

```
def make_table(m):
  tmpl = '{:>5}' *  (m+2)
  header = ['a']
  for i in range(1, m+2):
    header.append(i)
  print(tmpl.format(*header))
  for a in range(m):
    vals = [a]
    for i in range(1, m+2):
      vals.append(a**i % m)
    print(tmpl.format(*vals))

make_table(5)
```

```
def fermats_little_theorem(n, p):
  while True:
    a, b = divmod(n, p)
    if b == 1:
      break
    else:
      n, p = a, b
      a, b = divmod(n, p)
  return a
```

```
fermats_little_theorem(2**100, 5800)
```

```
def a_k_mod_m(a, k, m): #aのk乗をで割った余り
  b = 1
  for i in reversed(bin(k)[2:]):
    if i == '1':
      b = a * b % m
    a == a**2 % m
  return b
```

```
a_k_mod_m(7, 43, 123)
```

```
a_k_mod_m(2, pow(2, 100), 5801)
```
