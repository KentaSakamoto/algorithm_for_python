```
def merge_arrays(left, right = []):
  res = []
  i, j = 0, 0
  n, m = len(left), len(right)
  # どちらかの配列を調べ尽くしたらそこで終了
  while i < n and j < m:
    if left[i] < right[j]:
      res.append(left[i])
      i += 1
    else:
      # res.appned(right[j])
      res.append(right[j])
      j += 1
  # 残りはそのまま後ろに連結する
  return res + left[i:] + right[j:]
```

```
def step(array):
  res = []
  for i in range(0, len(array), 2):
    # 長さ2もしくは1の配列がスライスの結果を返る
    res.append(merge_arrays(*array[i:i+2]))
  return res
```

```
def merge_sort(array):
  if len(array) <= 1:
    return array
  mid_idx = len(array) // 2
  left = array[:mid_idx]
  right = array[mid_idx:]
  return merge_arrays(merge_sort(left), merge_sort(right))
```

```
def quick_sort(array):
  # 空の配列はそのまま返す
  if array == []:
    return array
  
  # 最後の要素をpivotにする
  p = array[-1]
  left = []
  right = []
  pivots = []

  # pivotとの関係で要素を分割する
  for v in array:
    if v < p:
      left.append(v)
    elif v == p:
      pivots.append(v)
    else:
      right.append(v)
  # 左と右は再び同じ関数を適用して返す
  return quick_sort(left) + pivots + quick_sort(right)
```
